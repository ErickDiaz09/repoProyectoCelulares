#Creaci√≥n de variables

##Se crea una variable presupesto con respecto al precio

```python
dif=df["price_usd"].max()-df["price_usd"].min()
pbajo=dif/3
print(df["price_usd"].min())
print(df["price_usd"].max())
print(f"Diferencia de rango de precio:{pbajo}")
print(f"Precio bajo <= {pbajo}")
print(f"Precio medio <= {pbajo*2}")
print(f"Precio alto > {pbajo*2}")
```

```python
def Budget(price):
    if price <= 466:
        return "Low"
    elif price <= 932:
        return "Mid"
    else:
        return "High"

df["budget"] = df["price_usd"].apply(Budget)
```

```python
df
```

##Se crea una variable con respecto al tama√±o de la pantalla

```python
dif=df["display_size_inch"].max()-df["display_size_inch"].min()
min=df["display_size_inch"].min()
tp=round(dif/3,2)
print(min)
print(df["display_size_inch"].max())
print(f"Diferencia de rango de precio:{dif}")
print(f"Tama√±o peque√±a <= {min+tp}")
print(f"Tama√±o mediana <= {min+(tp*2)}")
print(f"Tama√±o grande > {min+(tp*2)}")
```

```python
def Screen_size(display_size_inch):
    if display_size_inch <= 6.07:
        return "Small"
    elif display_size_inch <= 6.64:
        return "Medium"
    else :
        return "Big"
df["screen_size"] = df["display_size_inch"].apply(Screen_size)
```

# Visualizaciones

```python
plt.figure(figsize=(8,6))
df.boxplot(column="price_usd", by="os")
plt.xlabel("Sistema operativo")
plt.ylabel("Precio (USD)")
plt.title("Distribuci√≥n del precio por sistema operativo")
plt.suptitle("")
plt.tight_layout()
plt.show()
```

Ambas compa√±ias de sistema operativo no tiene diferencia en sus precios

```python
ram_mean = df.groupby("brand")["ram_gb"].mean().sort_values()
plt.figure(figsize=(12,5))
plt.bar(ram_mean.index, ram_mean.values)
plt.xticks(rotation=90)
plt.ylabel("RAM promedio (GB)")
plt.title("RAM promedio por marca")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("Global_Mobile_Prices_2025_Extended.csv")
group_counts = df.groupby(["os", "5g_support"]).size().unstack(fill_value=0)
plt.figure(figsize=(10,6))
group_counts.plot(kind="bar", stacked=True)
plt.xlabel("Sistema Operativo")
plt.ylabel("Cantidad de tel√©fonos")
plt.title("Distribuci√≥n de Soporte 5G por Sistema Operativo")
plt.legend(["No 5G", "S√≠ 5G"], title="5G Support")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
